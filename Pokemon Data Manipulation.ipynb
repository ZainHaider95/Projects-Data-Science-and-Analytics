```python
import pandas as pd
```

```python
data = pd.read_csv("Pokemon.csv")
data
```

# Data Manipulation / Feature Engineering

Adding High Low Attack Column

```python
HL_Attack = data["Attack"].mean()

def set_attack(val):
    if val < HL_Attack:
        return "Low Attack"
    elif val == HL_Attack:
        return "Average Attack"
    else:
        return "High Attack"
```

```python
data["HL_Attack"] = data["Attack"].apply(set_attack)
data
```

Adding High Low Speed Column

```python
HL_Speed = data["Speed"].mean()

def set_speed(val):
    if val < HL_Speed:
        return "Low Speed"
    elif val == HL_Speed:
        return "Average Speed"
    else:
        return "High Speed"
```

```python
data["HL_Speed"] = data["Speed"].apply(set_speed)
data
```

Adding High Low Speed Column

```python
HL_HP = data["HP"].mean()

def set_HP(val):
    if val < HL_HP:
        return "Low Horse Power"
    elif val == HL_HP:
        return "Average Horse Power"
    else:
        return "High Horse Power"
```

```python
data["HL_HP"] = data["HP"].apply(set_HP)
data
```
